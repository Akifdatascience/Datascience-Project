```python
#Checking for special character and replacing with null value
import re


def check_special_characters(df):
    special_characters = re.compile(r'[^a-zA-Z0-9 ]')
    has_special_characters = {}

    exclude_columns = ["BldgClassCategory"]

    for column in df.columns:
        if column in exclude_columns:
            has_special_characters[column] = False
            continue

        has_special_characters[column] = df[column].apply(
            lambda x: True if special_characters.search(str(x)) else False
        ).any()

        if has_special_characters[column]:
            df[column].replace(special_characters, np.nan, regex=True, inplace=True)

    return has_special_characters
```

```python
#Checking for negative value and converting into positive

def convert_negative_values_to_positive(df):
    numeric_columns = df.select_dtypes(include=np.number)

    for column in numeric_columns.columns:
        if (df[column] < 0).any():
            df[column] = df[column].abs()

    return df
```

```python
#checking for outliers and removing them

def replace_outliers_iqr(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    df = df.loc[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    return df
```

```python
import pandas as pd
import numpy as np

def replace_outliers_iqr(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    df.loc[df[column] > upper_bound, column] = upper_bound
    df.loc[df[column] < lower_bound, column] = lower_bound

    return df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score



# Create a Random Forest model object
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model to the training data
model.fit(x_train, y_train)

# Predict on the test data
y_pred = model.predict(x_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (R2):", r2)
```
